```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
def dXdt( t, X ) :                              
    return X[1], -1/4*X[1]-X[0]

# 초기값
X0 = ( 10, 0 ) 
t0 = 0
tend = 20

sol = solve_ivp( dXdt, (t0, tend), X0, t_eval=np.linspace(t0,tend,101) )
plt.plot( sol.t, sol.y[0] )
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
