```
import numba
from numba import jit
from numba import vectorize
import numpy as np
!find / -iname 'libdevice'
!find / -iname 'libnvvm.so'
import os
os.environ['NUMBAPRO_LIBDEVICE'] = "/usr/local/cuda-10.0/nvvm/libdevice"
os.environ['NUMBAPRO_NVVM'] = "/usr/local/cuda-10.0/nvvm/lib64/libnvvm.so"


x = int(input('몇 단을 원하세요?'))

def gogo_py(x):
    ans = np.zeros(10)
    for y in range(1, 10):
        ans[y - 1] = x * y
    return ans
print('기본 Python 버전')
ans = gogo_py(x)
for i in range(1, 10):
    print(x, '*', i, '=', ans[i - 1])
%timeit gogo_py(x)



def gogo_numpy(x):
    y = np.arange(1, 10)
    ans = np.multiply(x, y)
    return ans
print('Numpy 버전')
ans = gogo_numpy(x)
for i in range(1, 10):
    print(x, '*', i, '=', ans[i - 1])
%timeit gogo_numpy(x)



@jit
def gogo_jit(x):
    ans = np.zeros(10)
    for y in range(1, 10):
        ans[y - 1] = x * y
    return ans
print('JIT 버전')
ans = gogo_jit(x)
for i in range(1, 10):
    print(x, '*', i, '=', ans[i - 1])
%timeit gogo_jit(x)



@vectorize(['int64(int64)'], target='cuda')
def gogo_cuda(x):
    ans = np.zeros(10)
    for y in range(1, 10):
        ans[y - 1] = x * y
    return ans
print('Cuda 버전')
ans = int(gogo_cuda(x))
for i in range(1, 10):
    print(x, '*', i, '=', ans[i - 1])
%timeit gogo_cuda(x)
    
```
